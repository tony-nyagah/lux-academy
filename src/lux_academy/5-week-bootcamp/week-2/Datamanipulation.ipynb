## Measures of Central Tendency


Measures of central tendency are statistical metrics that describe the center of a data set. They include the mean, median, and mode.

Mean: The average of all data points.

Median: The middle value when data points are arranged in ascending or descending order.

Mode: The value that appears most frequently in the data set.

```python
data = [10, 20, 20, 30, 40]
mean_value = np.mean(data)
print(f"Mean: {mean_value}")
```

```python
median_value = np.median(data)
print(f"Median: {median_value}")
```

```python
data = [10, 20, 20, 30, 40]
data_series = pd.Series(data)

mode_value = data_series.mode()
print(f"Mode: {mode_value.values}")
```

### Univariate analysis

```python
df.head(3)
```

```python
df['Exited_Status'] = df['Exited'].map({0: 'stayed', 1: 'exited'})

# Get value counts of the new column
exited_counts = df['Exited_Status'].value_counts()

# Display the counts
print(exited_counts)
```

```python
#pie chart using matplotlib
exited = df["Exited_Status"].value_counts()

plt.figure(figsize=(12,6))
plt.pie(exited, labels=exited.index,autopct='%1.2f%%')
plt.title('Pie Chart for customers who exited')
plt.show()
```

```python
# Univariate Analysis - Age Distribution
sns.histplot(df['Age'], kde= True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Bivariate analysis

```python
#   tenure vs. Exited
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Tenure', hue='Exited_Status')
plt.title('Tenure vs. Exited')
plt.xlabel('Tenure')
plt.ylabel('Count')
plt.show()
```

```python
# Age vs. Balance
sns.scatterplot(data=df, x='Age', y='Balance')
plt.title('Age vs. Balance')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()
```

```python
df.corr(numeric_only='True')
```

## Multvariate analysis

```python
# Pair plot
sns.pairplot(data=df, vars=['Age', 'Balance', 'Exited'], hue='Exited')
plt.title('Pair Plot')
plt.show()
plt.tight_layout()
```

```python
# Select only numerical columns for correlation analysis
numerical_df = df.select_dtypes(include=['int64', 'float64'])

# Compute the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Correlation Matrix')
plt.show()
```
