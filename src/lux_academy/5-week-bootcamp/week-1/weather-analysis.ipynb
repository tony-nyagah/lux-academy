Answers to questions related to the weather dataset.

**Part 1**: From the dataset, write Python code to answer the following questions

```python
import pandas as pd

data = pd.read_csv("data/weather-data.csv")
data.head()
```

```python
data.tail()
```

```python
data.shape
```

1. Find all records where the weather was exactly clear.

```python
clear_skies = data[data["Weather"] == "Clear"]
clear_skies
```

2. Find the number of times the wind speed was exactly 4 km/hr.

```python
four_km_wind_speed = data[data["Wind Speed_km/h"] == 4]
four_km_wind_speed
```

3. Check if there are any NULL values present in the dataset.

```python
null_values = sum(data.isnull().values.ravel())
null_values
```

4. Rename the column "Weather" to "Weather_Condition."

```python
data.rename(columns={"Weather": "Weather_Condition"}, inplace=True)
data.head(2)
```

5. What is the mean visibility of the dataset?

```python
mean_visibility = data["Visibility_km"].mean()
mean_visibility
```

6. Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.

```python
wind_record = data[data["Wind Speed_km/h"] > 24]
wind_visibility_record = wind_record[wind_record["Visibility_km"] == 25]
wind_visibility_record
```

7. What is the mean value of each column for each weather condition?

```python
numerical_data = data[["Temp_C", "Dew Point Temp_C", "Rel Hum_%", "Wind Speed_km/h", "Visibility_km", "Press_kPa", "Weather_Condition"]]
mean_weather_condition_data = numerical_data.groupby("Weather_Condition").mean()
mean_weather_condition_data[:10]
```

8. Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.

```python
clear_weather_great_humidity_visibility = (data["Weather_Condition"] == "Clear") & ((data["Rel Hum_%"] > 50) | (data["Visibility_km"] > 40))
data[clear_weather_great_humidity_visibility][:10]
```

9. Find the number of weather conditions that include snow.

```python
snow_weather_conditions = data[data["Weather_Condition"].str.contains("Snow")]
len(snow_weather_conditions["Weather_Condition"].unique())
```

4. What is the meav visibility of the dataset?

```python
%%sql

SELECT AVG(Visibility_km) as mean_visibility 
FROM weather;
```

8. Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.

```python
%%sql

SELECT * 
FROM weather
WHERE Weather_Condition == 'Clear' AND ('Rel Hum_%' > 50 OR Visibility_km > 40)
LIMIT 5;
```

3. Check if there are any NULL values present in the dataset.

```python
%%sql

SELECT COUNT(*) as null_values
FROM weather
WHERE MAX(Temp_C, 'Dew Point Temp_C', 'Rel hum_%', 'Wind Speed_km/h', 'Visibility_km', 'Press_kPa', 'Weather_Condition') IS NULL;
```

4. Find the number of weather conditions that include snow.

```python
%%sql

SELECT DISTINCT COUNT(DISTINCT Weather_Condition) as snow_weather_conditions
FROM weather
WHERE Weather_Condition LIKE '%Snow%';
```
